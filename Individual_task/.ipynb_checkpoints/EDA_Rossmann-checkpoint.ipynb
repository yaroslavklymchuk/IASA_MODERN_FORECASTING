#### новогодние скачки продаж

```python
train_data.head()
```

```python
train_data['CompetitionOpenSinceDay'] = 1
```

```python
train_data.head()
```

```python
train_data['CompetitionOpenSinceDate'] = pd.to_datetime(dict(year=train_data.CompetitionOpenSinceYear,
                                                             month=train_data.CompetitionOpenSinceMonth,
                                                             day=train_data.CompetitionOpenSinceDay
                                                            )
                                                       )
```

```python
train_data.head()
```

```python
train_data['CompetitionOpenSinceDate'] = pd.to_datetime(train_data['CompetitionOpenSinceDate']).dt.date
```

```python
train_data.CompetitionOpenSinceDate.dropna().agg({min, max})
```

```python
train_data['qty_days_with_competitor'] = (train_data['Date'] - train_data['CompetitionOpenSinceDate']
                                         ).dt.days.fillna(0)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=train_data, x='StoreType', y='qty_days_with_competitor');
plt.title('Box-Plot of quantity days with competitor for each store type');
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=train_data, x='qty_days_with_competitor', y='Sales');
plt.title('Sales dependency over quantity of days with existed competitor');
```

```python
fig, axs = plt.subplots(nrows=train_data.StoreType.nunique(), ncols=1, figsize=(10, 30))

for i, store_type in enumerate(train_data.StoreType.unique()):
    sns.scatterplot(data=train_data, x='qty_days_with_competitor', y='Sales', ax=axs[i]);
    axs[i].set_title(
        f'Sales dependency over quantity of days with existed competitor for {store_type} store type');
```
